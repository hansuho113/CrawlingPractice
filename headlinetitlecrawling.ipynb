```python
import requests
from bs4 import BeautifulSoup
```

## 한 문장 뽑아오기

```python
url = 'https://news.naver.com/'    # url 주소 미리 지정
raw = requests.get(url)    # url 접속?
html = BeautifulSoup(raw.text, 'html.parser')    # html code parsing

hdlines = html.select('ul.hdline_article_list > li')    # "헤드라인 뉴스" 리스트 html로 접근
hdlines[0].select_one('a.lnk_hdline_article').text.strip()    # 첫 문장 뽑아서 양쪽 공백 제거
```

## For문 이용해서 헤드라인 뉴스 문장 5개 뽑아오기

```python
title_list = []
col_name =['Headlind News Title']

for hdline in hdlines:
    title = hdline.select_one('a.lnk_hdline_article').text.strip()
    title_list.append(title)   # title_list에 추가
df = pd.DataFrame(title_list, columns = col_name)    # DataFrame 변환
df
```
